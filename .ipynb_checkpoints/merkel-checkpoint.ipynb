```python
from lib import *
```

```python
air = AirFlow(temp=Q_(25, u.degC), pressure=Q_(1, u.atm), humidity=Q_(40,u.perc))
```

```python
water = WaterFlow(temp=Q_(40, u.degC), salinity=Q_(1000, u.mg/u.kg))
```

```python
solver = MerkelSolver(C=1.0, n=0.6, air_in=air, water_in = water, water_temp_out=Q_(30, u.degC))
```

```python
me = solver.solve_me(lg_ratio=1.0)
print(f"Merkel number = {me:.3f}")
```
